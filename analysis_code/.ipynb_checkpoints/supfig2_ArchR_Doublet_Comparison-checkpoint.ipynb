```r
#Utility functions

#extract cell barcodes to proj as i and s indices + soup_bc
extract_barcodes = function(proj, date_1){
    #extract raw barcode string + Subsample names
    x = t(data.frame(strsplit(proj$cellNames, '#'))[2,])
    x = data.frame(t(data.frame(strsplit(x, '_'))))
    
    #concact i7, i5 indices and add to proj
    proj$i = paste0(x[,1],'_', x[,2])
    
    #concact s7, s5 indices and add to proj
    proj$s = paste0(x[,3],'_', x[,4])   
    
    #donor
    proj$Donor = x[,9]
    
    #concact bed_bc and add to proj
    proj$bed_bc = paste0(x[,1],'_', x[,2], '-', x[,3],'_', x[,4])
    
    #concat soup_bc by date of run and add to proj
    proj$soup_bc = with(x, ifelse(X11 == date_1, 
                                  paste0(x[,1], x[,2],  x[,3], x[,4], '-1'), 
                                  paste0(x[,1], x[,2],  x[,3], x[,4], '-2')))
    
    return(proj)
}

#add_soup to proj as assignment
add_soup = function(proj, soup_dir){
    #import soup_data
    soup_data = read.csv(paste0(soup_dir, 'clusters.tsv'), sep='\t')

    #subset and sort soup_data by bc, then add to archr proj
    soup_data = soup_data[soup_data$barcode %in% proj$soup_bc,]
    soup_data = soup_data[match(proj$soup_bc, soup_data$barcode),]
    proj$assignment = soup_data$assignment
    proj$log_prob_singleton = soup_data$log_prob_singleton
    proj$log_prob_doublet = soup_data$log_prob_doublet
    proj$cluster0 = soup_data$cluster0
    proj$cluster1 = soup_data$cluster1
    
    return(proj)
}
```

```r
#load data
proj_name = 'DJ_Hs_PBMC_K27me3_220523_220527_79x79_MERGE_50k'
version = 'proj_dirty_final'
proj_dir <- paste0("/fh/fast/henikoff_s/user/jgreene/projects/sciCnT/data/ArchR/", proj_name, '/', version)
proj = loadArchRProject(path = proj_dir,
                 force = TRUE,
                 showLogo = TRUE)
setwd(proj_dir)
```

```r
#import souporcell calls
soup_dir = '/fh/fast/henikoff_s/user/jgreene/projects/sciCnT/data/DJ_Hs_PBMC_220523_79x79/souporcell/souporcell_DJ_Hs_PBMC_ALLSEQ_220523_220527_79x79_no_dup/'
proj = extract_barcodes(proj, '220523')
proj = add_soup(proj, soup_dir)
```

```r
#extract umap
umap_df = proj@cellColData %>% data.frame()
umap_df$UMAP_1 = proj@embeddings$UMAP$df[,1]
umap_df$UMAP_2 = proj@embeddings$UMAP$df[,2]

#reorder and plot
ord_new = c('0','1','Mixed')
umap_df$genotype <- factor(umap_df$genotype, levels = ord_new)
umap_df <- umap_df[order(umap_df$genotype),]

p0 = ggplot(umap_df, aes(x = UMAP_1, 
                    y = UMAP_2, 
                    color = genotype)
    )+geom_point(
    )+theme_void(
    )+theme(plot.title = element_blank(), 
            strip.text = element_blank(), 
            legend.position = c(.85, .85), 
            legend.text = element_text(size=25, family = 'Helvetica'),
            legend.title = element_text(size=25, family = 'Helvetica', face = "bold")
    )+guides(color = guide_legend(override.aes = list(size = 5))
    )+scale_color_manual(name = "Assignment",
                         values = c('#bfbfbf', '#7f7f7f', '#9705ff', '#3f00bd'
                                   )
)
p0
```
